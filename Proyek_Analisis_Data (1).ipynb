# Cleaning Data

```python
#mengubah nama-nama column pada day
day_df.rename(columns={'yr':'year','mnth':'month','weekday':'one_of_week', 'weathersit':'weather_situation', 'windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'},inplace=True)
```

```python
#mengubah nama-nama column pada hour
hour_df.rename(columns={'yr':'year','hr':'hours','mnth':'month','weekday':'one_of_week', 'weathersit':'weather_situation','windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'},inplace=True)
```

```python
#mengubah 1,2,3,4 menjadi 'spring', 'summer', 'fall', 'winter'
day_df.season.replace((1,2,3,4), ('Spring','Summer','Fall','Winter'), inplace=True)
hour_df.season.replace((1,2,3,4), ('Spring','Summer','Fall','Winter'), inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 : "Bagaimana distribusi jumlah sewa sepeda berdasarkan musim (season), dan musim mana yang memiliki jumlah sewa sepeda tertinggi?"

```python
# Menghitung jumlah sewa sepeda berdasarkan musim
sewa_per_musim = day_df.groupby('season')['count_cr'].sum().reset_index()

# Membuat plot
plt.figure(figsize=(10, 6))
plt.bar(sewa_per_musim['season'], sewa_per_musim['count_cr'], color='skyblue')
plt.title('Distribusi Jumlah Sewa Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Sewa Sepeda')
plt.xticks([0, 1, 2, 3], ['Spring', 'Summer', 'Fall', 'Winter'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Pertanyaan 2: "Apakah terdapat perbedaan signifikan dalam jumlah sewa sepeda antara musim-musim yang berbeda?"

```python
# Mengelompokkan data berdasarkan musim dan menghitung jumlah sewa sepeda
season_counts = day_df.groupby('season')['count_cr'].sum().reset_index()

# Membuat plot lingkaran
plt.figure(figsize=(8, 8))
plt.pie(season_counts['count_cr'], labels=season_counts['season'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Menambahkan judul
plt.title('Persentase Jumlah Sewa Sepeda Berdasarkan Musim')

# Menampilkan plot
plt.axis('equal')
plt.show()
```
